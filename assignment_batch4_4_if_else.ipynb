### 36. Determine the type of triangle using sides: acute, right, obtuse.

```python
a=int(input("Enter the 1st side of triangle:"))
b=int(input("Enter the 2st side of triangle:"))
c=int(input("Enter the 3st side of triangle:"))

if a+b+c !=180 or a <= 0 and b <= 0 and c <= 0:
    print(f"{a,b,c} plese check your angle size ** is incorrect")

elif a < 90 and b < 90 and c < 90:
    print(f"{a,b,c} its acute angle triangle")

elif a==90 or b==90 or c==90:
    print(f"{a,b,c} its right angle triangle")

elif a > 90 or b > 90 or c > 90:
    print(f"{a,b,c} its obtuse angle triangle")
```
